### Step 5 Compare 2 algorithms 
1) Use RandomForest for Classification task (Metrics: Acc, Confusion Matrix, ROC AUC)<br>
2) Use Naive Bayes for Classification task (Metrics: Acc, Confusion Matrix, ROC AUC)

```python
#Imports 

import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
from sklearn.model_selection import cross_val_score, KFold, cross_val_predict, StratifiedKFold

# Switch off warnings of Anaconda
import warnings
warnings.filterwarnings('ignore')
```

## RandomForestClassifier

```python
print('''Split data for fitting(lots of ways(Validation, CrossValidation))
  Straight solution with few models, also will define the benchmark''')

print('''\nDefine the Kfold CrossValidation''')
kf = KFold(n_splits=5, random_state=42, shuffle=True)
print(kf)


########################################## RandomForestClassifier MODEL #####################################################
print('\nRandomForestClassifier')
print('''\n For RandomForestClassifierwe will validate on 100 different DecisionTrees with max_depth = 2''')

def get_auc_rnd_forest_valid(X, y, seed = 17):
    #%time
    rand_forest = RandomForestClassifier(n_estimators = 100, max_depth = 2, random_state = seed)
    estimator = rand_forest
    prediction = cross_val_score(estimator, X, y, scoring='roc_auc', cv = kf)
    return(prediction.mean())


print('\n', 'RandomForestClassifier: ', get_auc_rnd_forest_valid(X_train, y_train, seed = 17))
```

```python
# #Executed №1 Random Forest
# print('''Split data for fitting(lots of ways(Validation, CrossValidation))
#   Straight solution with few models, also will define the benchmark''', '\n')

# print('''Define the Kfold CrossValidation''', '\n')
# kf = KFold(n_splits = 5, random_state = 17, shuffle = True)
# print(kf)


########################################## RandomForestClassifier MODEL #####################################################
print('\nRandomForestClassifier')

def get_auc_rnd_forest_valid(X, y, seed = 17):
    #%time
    rand_forest = RandomForestClassifier(random_state = 17, n_jobs = -1, oob_score = True)
    # Fit on Train Data
    results = cross_val_score(rand_forest, X, y, scoring = 'roc_auc', cv = skf)    
    # Check the accuracy on Test Data
    return("ROC AUC score: ", results.mean()*100)

print('\n', 'RandomForestClassifier: ', get_auc_rnd_forest_valid(X_train, y_train, seed = 17))
```

```python
# #Executed Random Forest Conf Matrix
########################################## RandomForestClassifier MODEL #####################################################
print('\nRandomForestClassifier')

def get_auc_rnd_forest_valid(X, y, seed = 17):
    #%time
    rand_forest = RandomForestClassifier(random_state = 17, n_jobs = -1, oob_score = True)
    y_pred = cross_val_predict(rand_forest, X, y, cv = skf)
    conf_matrix = confusion_matrix(y, y_pred)
    return("Confusion Matrix: ", conf_matrix)

print('\n', 'RandomForestClassifier: ', get_auc_rnd_forest_valid(X_train, y_train, seed = 17))
```

```python
#Executed №2 Random Forest
# Validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Create lists for saving accuracy both on Train and Test Data
train_acc = []
test_acc = []
temp_train_acc = []
temp_test_acc = []
trees_grid = [5, 10, 15, 20, 30, 50, 75, 100]

# Fit on Train Data
for ntrees in trees_grid:
    rfc = RandomForestClassifier(n_estimators=ntrees, random_state=42, n_jobs=-1, oob_score=True)
    temp_train_acc = []
    temp_test_acc = []
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        rfc.fit(X_train, y_train)
        temp_train_acc.append(rfc.score(X_train, y_train))
        temp_test_acc.append(rfc.score(X_test, y_test))
    train_acc.append(temp_train_acc)
    test_acc.append(temp_test_acc)
    
train_acc, test_acc = np.asarray(train_acc), np.asarray(test_acc)
print("Best accuracy is {:.2f}% with {} trees".format(max(test_acc.mean(axis=1))*100, 
                                                        trees_grid[np.argmax(test_acc.mean(axis=1))]))
```

### Naive Biase

```python
#Define and fit Naive Bayes Model 
model = GaussianNB().fit(X_train, y_train)

#Make Predictions
predicted = model.predict(X_test)

#Calculate ROC 
fpr, tpr, thresholds = roc_curve(y_test, predicted)
##print(fpr, tpr, thresholds)

#Calculate AUC
auc = roc_auc_score(y_test, predicted)
print('Naive Baise AUC: %.3f' % auc)

#Confusion Matrix
print('Naive Baise Confusion Matrix: ', '\n' ,confusion_matrix(y_test, predicted))


#Plotting the results
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
# show the plot
plt.show()
```
